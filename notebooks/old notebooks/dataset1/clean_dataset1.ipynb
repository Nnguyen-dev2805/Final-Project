````python
# So sánh holiday/weather gốc của Dataset 1 với dữ liệu enrich từ calendar + Open-Meteo

compare_df = df_raw.copy()

compare_df["date"] = pd.to_datetime(compare_df["date"], errors="coerce").dt.normalize()

# Chuẩn hóa holiday gốc
compare_df["original_holiday_name_clean"] = (
    compare_df["original_holiday_name"]
    .fillna("No Holiday")
    .astype(str)
    .str.strip()
)

compare_df["calendar_holiday_name_clean"] = (
    compare_df["holiday_name"]
    .fillna("No Holiday")
    .astype(str)
    .str.strip()
)

compare_df["original_is_holiday"] = compare_df["original_holiday_name_clean"].ne("No Holiday")
compare_df["calendar_is_holiday"] = compare_df["calendar_holiday_name_clean"].ne("No Holiday")

# Chuẩn hóa weather gốc
compare_df["original_temperature_c"] = pd.to_numeric(
    compare_df["original_temperature_c"],
    errors="coerce"
)

compare_df["temperature_c"] = pd.to_numeric(
    compare_df["temperature_c"],
    errors="coerce"
)

compare_df["temperature_diff"] = (
    compare_df["temperature_c"] - compare_df["original_temperature_c"]
)

compare_df["abs_temperature_diff"] = compare_df["temperature_diff"].abs()

compare_df["original_weather_condition_clean"] = (
    compare_df["original_weather_condition"]
    .fillna("Missing")
    .astype(str)
    .str.strip()
)

compare_df["open_meteo_weather_condition_clean"] = (
    compare_df["weather_condition"]
    .fillna("Missing")
    .astype(str)
    .str.strip()
)

compare_df["weather_condition_match"] = (
    compare_df["original_weather_condition_clean"]
    == compare_df["open_meteo_weather_condition_clean"]
)

compare_df["holiday_match"] = (
    compare_df["original_holiday_name_clean"]
    == compare_df["calendar_holiday_name_clean"]
)

print("===== Holiday comparison =====")
print("Rows:", len(compare_df))
print("Original holiday rows:", compare_df["original_is_holiday"].sum())
print("Calendar holiday rows:", compare_df["calendar_is_holiday"].sum())
print("Exact holiday name match:", compare_df["holiday_match"].mean() * 100)

display(
    pd.crosstab(
        compare_df["original_is_holiday"],
        compare_df["calendar_is_holiday"],
        rownames=["Original is holiday"],
        colnames=["Calendar is holiday"],
        margins=True
    )
)

print("\nHoliday mismatches:")
holiday_mismatch = compare_df[
    ~compare_df["holiday_match"]
][
    [
        "date",
        "timestamp",
        "original_holiday_name_clean",
        "calendar_holiday_name_clean",
        "is_holiday",
        "holiday_type",
        "is_official_holiday",
        "is_commercial_event",
    ]
].drop_duplicates()

display(holiday_mismatch.head(50))
print("Number of unique holiday mismatch dates:", holiday_mismatch["date"].nunique())


print("\n===== Temperature comparison =====")
display(
    compare_df[
        [
            "original_temperature_c",
            "temperature_c",
            "temperature_diff",
            "abs_temperature_diff",
        ]
    ].describe()
)

print("Missing original_temperature_c:", compare_df["original_temperature_c"].isna().sum())
print("Missing Open-Meteo temperature_c:", compare_df["temperature_c"].isna().sum())

print("\nRows with abs temperature difference >= 5C:")
display(
    compare_df[
        compare_df["abs_temperature_diff"] >= 5
    ][
        [
            "timestamp",
            "date",
            "hour",
            "original_temperature_c",
            "temperature_c",
            "temperature_diff",
            "original_weather_condition_clean",
            "open_meteo_weather_condition_clean",
        ]
    ].head(50)
)


print("\n===== Weather condition comparison =====")
print("Weather condition exact match:", compare_df["weather_condition_match"].mean() * 100)

display(
    pd.crosstab(
        compare_df["original_weather_condition_clean"],
        compare_df["open_meteo_weather_condition_clean"],
        rownames=["Original weather"],
        colnames=["Open-Meteo weather"],
        margins=True
    )
)

print("\nWeather condition mismatches:")
display(
    compare_df[
        ~compare_df["weather_condition_match"]
    ][
        [
            "timestamp",
            "date",
            "hour",
            "original_weather_condition_clean",
            "open_meteo_weather_condition_clean",
            "original_temperature_c",
            "temperature_c",
            "abs_temperature_diff",
        ]
    ].head(50)
)
````

### 2. Helper functions

````python
# chuyển một cột dữ liệu về kiểu Boolean
def to_bool(series, default=False):
    mapping = {
        'true': True,
        'false': False,
        '1': True,
        '0': False,
        'yes': True,
        'no': False,
        'y': True,
        'n': False,
    }
    if series.dtype == bool:
        return series.fillna(default).astype(bool)
    return series.astype(str).str.strip().str.lower().map(mapping).fillna(default).astype(bool)

# lấy giá trị xuất hiện nhiều nhất trong một cột dữ liệu
def first_mode(series):
    non_missing = series.dropna()
    if non_missing.empty:
        return pd.NA
    mode_values = non_missing.mode()
    return mode_values.iloc[0] if not mode_values.empty else non_missing.iloc[0]

# tính ngưỡng phát hiện outlier bằng phương pháp IQR
def iqr_bounds(series, multiplier=1.5):
    q1 = series.quantile(0.25)
    q3 = series.quantile(0.75)
    iqr = q3 - q1
    return q1 - multiplier * iqr, q3 + multiplier * iqr
````

### 3. Chuẩn hóa schema transaction-level

Dataset 1 chỉ có địa điểm ở cấp city. Để dùng chung schema với Dataset 2, ta tạo:

```text
city = New York
area = city_level_unknown
location_level = city
```

````python
df = df_raw.copy()

df['source_dataset'] = 'dataset1'
df['transaction_datetime'] = pd.to_datetime(
    df['timestamp_datetime'] if 'timestamp_datetime' in df.columns else df['timestamp'],
    errors='coerce',
)
df['date'] = df['transaction_datetime'].dt.normalize()
df['month'] = df['transaction_datetime'].dt.month
df['weekday'] = df['transaction_datetime'].dt.day_name()
df['weekday_num'] = df['transaction_datetime'].dt.dayofweek
df['hour'] = df['transaction_datetime'].dt.hour
df['is_weekend'] = df['weekday_num'].isin([5, 6])

city_candidates = [col for col in ['city_y', 'city_x', 'city'] if col in df.columns]
df['city'] = df[city_candidates[0]] if city_candidates else 'New York'
for col in city_candidates[1:]:
    df['city'] = df['city'].fillna(df[col])
df['city'] = df['city'].fillna('New York')
df['area'] = 'city_level_unknown'
df['location_level'] = 'city'
df['location'] = df['city']
df['store_id'] = pd.NA
df['store_location'] = df['city']

df['product_category'] = df['product_category'].astype(str).str.strip()
df['product_detail_clean'] = df['product_name'].astype(str).str.strip() if 'product_name' in df.columns else pd.NA

df['unit_price'] = pd.to_numeric(df['unit_price'], errors='coerce')
df['quantity'] = pd.to_numeric(df['quantity'], errors='coerce')
df['revenue'] = pd.to_numeric(df['total_amount'], errors='coerce')
df['transaction_count'] = 1

numeric_external_cols = [
    'temperature_c',
    'apparent_temperature_c',
    'relative_humidity',
    'precipitation_mm',
    'rain_mm',
    'weather_code',
    'wind_speed_10m_kmh',
    'cloud_cover_pct',
    'days_to_nearest_holiday',
]
for col in numeric_external_cols:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce')

bool_cols = [
    'is_rainy',
    'is_snowy',
    'is_precipitation',
    'is_holiday',
    'is_official_holiday',
    'is_commercial_event',
    'is_pre_holiday',
    'is_post_holiday',
    'discount_applied',
    'loyalty_member',
]
for col in bool_cols:
    if col in df.columns:
        df[col] = to_bool(df[col])

common_transaction_cols = [
    'source_dataset',
    'transaction_id',
    'transaction_datetime',
    'date',
    'month',
    'weekday',
    'weekday_num',
    'hour',
    'is_weekend',
    'city',
    'area',
    'location_level',
    'location',
    'store_id',
    'store_location',
    'product_category',
    'product_detail_clean',
    'unit_price',
    'quantity',
    'transaction_count',
    'revenue',
    'temperature_c',
    'apparent_temperature_c',
    'relative_humidity',
    'precipitation_mm',
    'rain_mm',
    'weather_condition',
    'weather_code',
    'wind_speed_10m_kmh',
    'cloud_cover_pct',
    'is_rainy',
    'is_snowy',
    'is_precipitation',
    'weather_spatial_level',
    'is_holiday',
    'holiday_name',
    'holiday_type',
    'is_official_holiday',
    'is_commercial_event',
    'is_pre_holiday',
    'is_post_holiday',
    'days_to_nearest_holiday',
    'nearest_holiday_name',
]

df1_clean = df[common_transaction_cols].copy()

print('Dataset 1 clean transaction-level shape:', df1_clean.shape)
display(df1_clean.head())
````

### 5. Aggregate về cấp mô hình

Không nên train ở transaction-level nếu mục tiêu là dự đoán doanh thu theo external factors. Cấp hợp lý hơn:

```text
date + hour + city + area + product_category
```

Dataset 1 có `area = city_level_unknown`, vì dữ liệu chỉ ở cấp thành phố.

````python
# Tổng hợp doanh thu theo từng giờ, từng ngày, từng khu vực, từng loại sản phẩm

group_cols_total = [
    'source_dataset',
    'date',
    'month',
    'weekday',
    'weekday_num',
    'hour',
    'is_weekend',
    'city',
    'area',
    'location_level',
]

df1_hourly_total = (
    df1_clean.groupby(group_cols_total, dropna=False)
    .agg(
        revenue=('revenue', 'sum'),
        quantity=('quantity', 'sum'),
        transaction_count=('transaction_count', 'sum'),
        avg_transaction_value=('revenue', 'mean'),
        avg_unit_price=('unit_price', 'mean'),

        temperature_c=('temperature_c', 'mean'),
        apparent_temperature_c=('apparent_temperature_c', 'mean'),
        relative_humidity=('relative_humidity', 'mean'),
        precipitation_mm=('precipitation_mm', 'mean'),
        wind_speed_10m_kmh=('wind_speed_10m_kmh', 'mean'),
        cloud_cover_pct=('cloud_cover_pct', 'mean'),

        weather_condition=('weather_condition', first_mode),
        is_rainy=('is_rainy', first_mode),
        is_snowy=('is_snowy', first_mode),
        is_precipitation=('is_precipitation', first_mode),
        weather_spatial_level=('weather_spatial_level', first_mode),

        is_holiday=('is_holiday', first_mode),
        holiday_name=('holiday_name', first_mode),
        holiday_type=('holiday_type', first_mode),
        is_official_holiday=('is_official_holiday', first_mode),
        is_commercial_event=('is_commercial_event', first_mode),
        is_pre_holiday=('is_pre_holiday', first_mode),
        is_post_holiday=('is_post_holiday', first_mode),
        days_to_nearest_holiday=('days_to_nearest_holiday', 'min'),
        nearest_holiday_name=('nearest_holiday_name', first_mode),
    )
    .reset_index()
)

round_cols = [
    'revenue',
    'avg_transaction_value',
    'avg_unit_price',
    'temperature_c',
    'apparent_temperature_c',
    'precipitation_mm',
]

for col in round_cols:
    df1_hourly_total[col] = df1_hourly_total[col].round(3)

print('Dataset 1 hourly total aggregate shape:', df1_hourly_total.shape)
display(df1_hourly_total.head())
display(df1_hourly_total['revenue'].describe())
       
````
